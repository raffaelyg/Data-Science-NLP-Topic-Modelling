**Reasoning**:
The preprocessing steps, including tokenization, have been applied. Now, as per the project guidance, I need to find the frequency distribution of words from each data set's cleaned reviews separately using `nltk.FreqDist`.



```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
import re

# Download necessary NLTK data
nltk.download('stopwords')
nltk.download('punkt')

# Define preprocessing function
def preprocess_text(text):
    if not isinstance(text, str):
        return ""
    # Lowercase
    text = text.lower()
    # Remove numbers
    text = re.sub(r'\\d+', '', text)
    # Tokenize
    tokens = word_tokenize(text)
    # Remove stopwords
    stop_words = set(stopwords.words('english'))
    tokens = [word for word in tokens if word.isalpha() and word not in stop_words]
    # Join back into a string for FreqDist
    return ' '.join(tokens)

# Apply preprocessing to Google reviews
google_df['cleaned_comment'] = google_df['Comment'].apply(preprocess_text)

# Apply preprocessing to Trustpilot reviews
trustpilot_df['cleaned_review_content'] = trustpilot_df['Review Content'].apply(preprocess_text)

print("Text preprocessing complete. New 'cleaned_comment' and 'cleaned_review_content' columns created.")

from nltk.probability import FreqDist

# Get all words from Google reviews
google_words = ' '.join(google_df['cleaned_comment']).split()
google_freq_dist = FreqDist(google_words)

# Get all words from Trustpilot reviews
trustpilot_words = ' '.join(trustpilot_df['cleaned_review_content']).split()
trustpilot_freq_dist = FreqDist(trustpilot_words)

print("Frequency distribution for Google reviews (top 10):")
print(google_freq_dist.most_common(10))

print("\nFrequency distribution for Trustpilot reviews (top 10):")
print(trustpilot_freq_dist.most_common(10))
```

**Reasoning**:
I need to visualize the top 10 most frequent words from each dataset. I will ensure the plots have appropriate titles, labels, and rotations for readability.



```python
import matplotlib.pyplot as plt
import seaborn as sns

# Get top 10 words for Google reviews
google_top_10_words = google_freq_dist.most_common(10)
google_words_df = pd.DataFrame(google_top_10_words, columns=['Word', 'Frequency'])

# Get top 10 words for Trustpilot reviews
trustpilot_top_10_words = trustpilot_freq_dist.most_common(10)
trustpilot_words_df = pd.DataFrame(trustpilot_top_10_words, columns=['Word', 'Frequency'])

# Plotting for Google reviews
plt.figure(figsize=(12, 6))
sns.barplot(x='Word', y='Frequency', data=google_words_df, palette='viridis')
plt.title('Top 10 Most Frequent Words in Google Reviews (After Preprocessing)')
plt.xlabel('Words')
plt.ylabel('Frequency')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# Plotting for Trustpilot reviews
plt.figure(figsize=(12, 6))
sns.barplot(x='Word', y='Frequency', data=trustpilot_words_df, palette='magma')
plt.title('Top 10 Most Frequent Words in Trustpilot Reviews (After Preprocessing)')
plt.xlabel('Words')
plt.ylabel('Frequency')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

## Retry: Load Falcon Model and Extract Topics (GPU check)

### Subtask:
Load the `tiiuae/falcon-7b-instruct` model and apply the specified prompt to extract the top 3 topics from each negative review, handling potential long execution times by using a subset if necessary, after ensuring a GPU runtime is active.


```python
import re

extracted_topics_per_review = []
prompt_prefix = "In the following customer review, pick out the main 3 topics. Return them in a numbered list format, with each one on a new line."

# Note: This process can be time-consuming even for a subset.
for i, review in enumerate(data_subset):
    # Construct the full prompt for the current review
    full_prompt = f"{prompt_prefix}\n\nReview: {review}\nTopics:"
    try:
        # Generate text using the pipeline with appropriate parameters
        # max_new_tokens limits the length of the generated topics
        # do_sample and top_k are for more diverse generation
        output = pipeline_generator(
            full_prompt,
            max_new_tokens=100, # Max tokens for the generated topics (adjust as needed)
            do_sample=True,
            top_k=10,
            num_return_sequences=1,
            eos_token_id=pipeline_generator.tokenizer.eos_token_id, # Ensure generation stops at EOS token
            pad_token_id=pipeline_generator.tokenizer.eos_token_id # Use eos_token_id for pad_token_id
        )
        generated_text = output[0]['generated_text']

        # Extract only the part after "Topics:" as the generated response
        topics_part = generated_text.split("Topics:")[-1].strip()

        # Parse the numbered list of topics
        topics = []
        for line in topics_part.split('\n'):
            match = re.match(r'^\s*\d+\.\s*(.*)', line)
            if match:
                topic = match.group(1).strip()
                if topic: # Ensure topic is not empty
                    topics.append(topic)
        extracted_topics_per_review.append(topics)

    except Exception as e:
        print(f"Error processing review {i}: {e}")
        extracted_topics_per_review.append([]) # Append an empty list or error indicator for failed reviews

print(f"Successfully extracted topics for {len(extracted_topics_per_review)} reviews.")
print("\nFirst 3 extracted topic lists (for the subset):")
for i in range(min(3, len(extracted_topics_per_review))):
    print(f"Review {i}: {extracted_topics_per_review[i]}")
```

```python
import gensim
from gensim import corpora

# 1. Tokenize each review in the all_negative_reviews list
# Each review is already a string of space-separated words from previous preprocessing steps,
# so we can simply split each string into a list of words.
tokenized_reviews = [review.split() for review in all_negative_reviews]

# 2. Create a Gensim Dictionary from the tokenized_reviews
dictionary = corpora.Dictionary(tokenized_reviews)

# 3. Create a Bag-of-Words (BoW) corpus
corpus = [dictionary.doc2bow(review) for review in tokenized_reviews]

print(f"Number of tokenized reviews: {len(tokenized_reviews)}")
print(f"Number of unique tokens in dictionary: {len(dictionary)}")
print(f"Number of documents in corpus: {len(corpus)}")
print("First 5 tokenized reviews:")
for i in range(min(5, len(tokenized_reviews))):
    print(f"- {tokenized_reviews[i][:10]}...") # Print first 10 words for brevity
print("\nFirst 5 entries in the corpus (first review's BoW representation):")
if corpus:
    print(f"- {corpus[0][:5]}...") # Print first 5 (word_id, count) tuples for brevity
```

## Perform LDA on tokenised data

### Subtask:
Using Gensim, perform LDA on the tokenised data, specifying the number of topics as 10.


**Reasoning**:
I need to initialize and train a Gensim LDA model on the previously prepared corpus and dictionary, specifying 10 topics. After training, I will extract and print the top 10 most contributing words for each of these 10 topics.



```python
from gensim.models import LdaModel

# Set number of topics
num_topics = 10

# Initialize LDA model
# 'passes' parameter controls how many times the training algorithm will iterate over the entire corpus.
# 'random_state' ensures reproducibility.
lda_model = LdaModel(corpus=corpus, id2word=dictionary, num_topics=num_topics, passes=15, random_state=100)

# Print the top 10 most contributing words for each topic
print("LDA Model Topics:")
for idx, topic in lda_model.print_topics(num_words=10):
    print(f"Topic {idx}: {topic}")
```

### Comment on LDA Output and Comparison

The `pyLDAvis` visualization successfully generated an interactive representation of the 10 topics identified by Gensim's LDA model. This visualization allows for exploration of inter-topic distance maps and the salient terms within each topic.

**Comparison with BERTopic (Initial Run on Negative Reviews):**

- **Topic Interpretation**: Both LDA and BERTopic aim to uncover underlying themes. Comparing the top words from LDA topics to those identified by BERTopic in the initial run (on negative reviews from common locations) reveals some parallels but also differences in granularity and focus.
    - For instance, LDA Topic 0 (`gym`, `equipment`, `machines`, `people`, `use`, `one`, `time`, `staff`, `weights`) seems to broadly cover issues related to **gym usage, equipment, and overcrowding**, which aligns with several of BERTopic's negative themes like 'Equipment Condition/Availability' and general 'Gym experience/Crowding'.
    - LDA Topic 5 (`gym`, `water`, `showers`, `broken`, `hot`, `air`, `shower`, `changing`, `cold`, `dirty`) strongly reflects **cleanliness and facilities issues**, specifically `showers` and `water` temperature, similar to BERTopic's 'Cleanliness of Facilities' and 'Air Conditioning/Temperature Issues'.
    - LDA Topic 3 (`gym`, `parking`, `closed`, `open`, `car`, `pay`, `park`, `free`, `membership`, `hours`) clearly points to **parking and operational hours/access problems**, directly corresponding to BERTopic's 'Parking Issues' and 'Opening/Closing Hours & Access'.
    - LDA Topic 6 (`membership`, `gym`, `get`, `month`, `service`, `app`, `cancel`, `customer`, `email`, `account`) directly addresses **membership and cancellation issues**, a recurring theme also found in BERTopic.
    - LDA Topic 8 (`gym`, `air`, `conditioning`, `pt`, `issues`, `turn`, `member`, `temperature`, `working`, `staff`) is highly focused on **air conditioning/temperature issues and personal training (PT)**, which was a distinct topic in BERTopic.

- **Multilingual Topics**: A notable difference in the LDA output is Topic 4 (`er`, `og`, `det`, `der`, `ikke`, `på`, `de`, `man`, `jeg`, `har`), which appears to consist of Danish words. This indicates that the preprocessing for Gensim's LDA (which was applied to `all_negative_reviews`) did not specifically filter out non-English stopwords or reviews in other languages, unlike the BERTopic preprocessing which was more general.

- **Topic Granularity**: BERTopic often provides more coherent and distinct topics due to its use of BERT embeddings, which capture semantic similarity better than traditional bag-of-words models like LDA. However, LDA still effectively highlights major discussion points when given a sufficiently cleaned corpus.

**Extra Insights Obtained:**

- **Validation**: The LDA results largely validate the findings from BERTopic regarding the main areas of negative feedback. Both models highlight similar pain points such as equipment, cleanliness, staff, membership, parking, and climate control.
- **Language Diversity**: The identification of a foreign language topic in LDA is an interesting insight, suggesting that a portion of the negative reviews are not in English. This could indicate a need for more robust multilingual preprocessing or targeted analysis for non-English reviews if the PureGym Group operates in countries where reviews are submitted in local languages.

In conclusion, while both models provide valuable insights, BERTopic typically offers a more semantically rich and often clearer separation of topics, while LDA provides a good, interpretable baseline. The consistency in identifying major themes across both techniques strengthens the confidence in these findings.
